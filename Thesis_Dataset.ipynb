```python
import pandas as pd
from thefuzz import process

# Load datasets
wta_player_path = "Output/WTA_match_final_file.csv"
wta_prize_money_path = "wta_prize_money_clean.csv"

wta_player = pd.read_csv(wta_player_path)
wta_prize_money = pd.read_csv(wta_prize_money_path)

#Function to standardize player names
def standardize_name(name):
    if isinstance(name, str):
        name = name.lower().strip()  # Convert to lowercase & remove extra spaces
        name = name.replace(",", "").replace(".", "").replace("-", " ")  # Remove punctuation & normalize spacing
        return " ".join(name.split())  # Ensure single spaces only
    return name

# Apply standardization to both datasets
wta_player['player_name'] = wta_player['player_name'].apply(standardize_name)
wta_prize_money['player_name'] = wta_prize_money['player_name'].apply(standardize_name)

# Identify unmatched player names
players_not_found = set(wta_player['player_name']) - set(wta_prize_money['player_name'])
print(f"{len(players_not_found)} player names not found in prize money dataset")
print("Example names:", list(players_not_found)[:10])

# Fuzzy match player names for better merging
matched_names = {}
for name in players_not_found:
    match, score = process.extractOne(name, wta_prize_money['player_name'])
    if score > 85:  # Only consider high-confidence matches
        matched_names[name] = match

print(f"{len(matched_names)} player names matched using fuzzy matching")

# Replace player names in wta_player with best fuzzy match
wta_player['player_name'] = wta_player['player_name'].replace(matched_names)

# Merge the datasets on 'player_name'
merged_data = pd.merge(wta_player, wta_prize_money, on='player_name', how='left')

# Ensure prize money columns exist before filling missing values
for col in ['singles', 'doubles', 'total']:
    if col in merged_data.columns:
        merged_data[col] = merged_data[col].fillna(0)
    else:
        print(f"⚠️ Warning: Column '{col}' not found in prize money dataset!")

# Save the merged dataset
output_path = "Output/Wta_player_with_prizemoney.csv"
merged_data.to_csv(output_path, index=False)

print("Merged dataset with prize money saved successfully!")
print("Sample data preview:")
print(merged_data.head())
```
